```python
from statsmodels.tsa.stattools import adfuller
def adf(col):
    result = adfuller(col)
    print(result)
    if result[1]<0.05:
        print(f'{col.name} is stationary')
    else:
        print(f'{col.name} is non-stationary')
```

```python
adf(df8['WSR0'])
adf(df8['WSR1'])
```

# Vector Autoregressor

```python
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.statespace.varmax import VARMAX
from sklearn.model_selection import train_test_split
```

```python
def gran(col1,col2):
    print(f'{col1} causes {col2}')
    print('-----------------------------------------------------------------------------------')
    granger = grangercausalitytests(df8[[col1,col2]],2)
```

```python
gran('WSR0','WSR1')
```

```python
gran('WSR1','WSR0')
```

```python
train,test,_,_ = train_test_split(df8[['WSR0','WSR1']],df8['O3'],test_size=0.2,shuffle=False)
train,test
```

```python
from statsmodels.tsa.api import VAR
model = VAR(train)
sortedmodel = model.select_order(maxlags=20)
print(sortedmodel.summary())
```

```python
var_model = VARMAX(train,order=(10,0),enforce_stationarity=True)
fitmodel = var_model.fit()
fitmodel.summary()
```

```python
n_forecast = 12
pred = fitmodel.get_prediction()

preds = pred.predicted_mean
```

```python
preds.columns = ['WSR0 pred','WSR1 pred']
preds
```

```python
testvspred = pd.concat([train,preds],axis=1)
testvspred
```

```python
testvspred[['WSR0','WSR0 pred']].plot()
```

```python
testvspred[['WSR1','WSR1 pred']].plot()
```

```python
from sklearn.metrics import mean_squared_error as mse
mse(testvspred['WSR1'],testvspred['WSR1 pred'])
```

```python
mse(testvspred['WSR0'],testvspred['WSR0 pred'])
```
