```python
import pandas as pd
time_series=pd.date_range('2025-01-01','2025-01-03')
time_series=pd.DataFrame(data=time_series,columns=['timestamp'])
time_series.set_index('timestamp',inplace=True)

#performing upsampling
upsampled_data=time_series.resample('1min').asfreq()
upsampled_data
```

```python
#performing down sampling
downsampled_data=upsampled_data.resample('5min').asfreq()
downsampled_data
```

```python
from datetime import datetime
def parser(x):
    return datetime.strptime('190'+x,'%Y-%m') 
series=pd.read_csv(url,header=0,parse_dates=[0],index_col=0,date_parser=parser)
print(series)
```

```python
#upsampling using interpolation
upsampled=series.resample('D')
interpolated=upsampled.interpolate(method="linear")
interpolated
```

```python
#downsampling
resampled=series.resample('Q')
resampled=resampled.mean()
resampled
```
