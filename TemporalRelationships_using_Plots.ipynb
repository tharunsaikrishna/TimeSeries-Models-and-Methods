```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/admin/Downloads/daily-min-temperatures.csv")
df
```

```python
df.set_index('Date',inplace=True)
df
```

```python
plt.figure(figsize=(20,8))
plt.plot(df.index,df['Temp'])
plt.title('line plot')
plt.xlabel('Date')
plt.ylabel('Temp')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(df.index,df['Temp'])
plt.title('scatter plot')
plt.xlabel('Date')
plt.ylabel('Temp')
plt.show()
```

```python
plt.figure(figsize=(20,8))
pd.plotting.autocorrelation_plot(df['Temp'])
plt.title("Autocorrelation plot")
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.hist(df['Temp'],bins=100)
plt.title("histograms of temperatures")
plt.xlabel('Temp in celcius')
plt.ylabel('frequency')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.kdeplot(df['Temp'],fill=True)
plt.title("Density plot")
plt.xlabel('Temp in celcius')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(2,2))
sns.boxplot(x=df['Temp'])
plt.title('Box and Whisker plot')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df,cmap='coolwarm')
plt.title('Heat Map')
plt.xlabel('Temperature')
plt.show()
```
