```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("E:/126156163/jan 11/daily-min-temperatures.csv")
df.head(10)
```

```python
df['Temp'].plot(kind='hist',bins=20,title='Temperature')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
plt.show()
```

```python
df['Temp'].plot(kind='line',title='Temperature')
ax=plt.gca()
```

```python
df['lag_1']=df['Temp'].shift(1)
df['lag_2']=df['Temp'].shift(2)
```

```python
df.head()
```

```python
df['sliding_mean_3']=df['Temp'].rolling(3).mean()
df['sliding_std_3']=df['Temp'].rolling(3).std()
df['sliding_min_3']=df['Temp'].rolling(3).min()
df['sliding_max_3']=df['Temp'].rolling(3).max()
df
```

```python
df['expanding_mean']=df['Temp'].expanding().mean()
df['expanding_std']=df['Temp'].expanding().std()
df['expanding_min']=df['Temp'].expanding().min()
df['expanding_max']=df['Temp'].expanding().max()
df
```
